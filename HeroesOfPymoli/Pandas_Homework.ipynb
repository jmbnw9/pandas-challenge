# Purchase Data Analysis
A few trends that I observed in the data set are that the majority of players are male. They were completing 84% of the gathered data. Another trend is that the biggest percentage of players are in the age bracket of 20-24, making up roughly 45% of the data set. Lastly, I noticed that the item mostly frequently purchase were Final Critic and Oathbreaker. The other items took a decent drop in sales compared to these two.

```python
# Import Dependencies 
import pandas as pd
```

```python
#Store the file path 
file = "Resources/purchase_data.csv"
```

```python
# Read the file
purchase_data_df = pd.read_csv(file)
purchase_data_df.head()
```

```python
#Find the total number of players
total_players = len(purchase_data_df["SN"].unique())
```

```python
#Print the total players into a summary table
total_summary_df = pd.DataFrame({"Total Players": [total_players]})
                              
total_summary_df
```

```python
# Find the number of unique items, average price, number of purchases, and total revenue
unique_items = len(purchase_data_df["Item Name"].unique())

average = purchase_data_df["Price"].mean()

purchase_data_df["Purchase ID"].count()
purchase_number = purchase_data_df["Purchase ID"].count()

purchase_data_df["Price"].sum()
revenue = purchase_data_df["Price"].sum()
```

```python
# Print the summary table
summary_df = pd.DataFrame({"Number of Unique items": [unique_items], "Average Price": [average], "Number Of Purchases": [purchase_number], "Total Revenue": [revenue]})
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)
summary_df.head()
```

```python
#Group by gender demographics
grouped_gender_df = purchase_data_df.groupby(['Gender'])

#Find the total count per gender
total_count = grouped_gender_df.nunique()["SN"]

#Percentage per gender
player_percentage = total_count/total_players *100

#Create the summary table
summary_df = pd.DataFrame({"Total Count": total_count, "Percentage of Players": player_percentage})

summary_df.index.name = None

summary_df["Percentage of Players"] = summary_df["Percentage of Players"].map("{0:.2f}%".format)

summary_df.head()
```

```python
#Find the purchase count, average purchase price, total purchase value, and average total purchase per person
grouped_gender_df = purchase_data_df.groupby(['Gender'])

purchase_count = grouped_gender_df["Purchase ID"].count()

average_purchase = grouped_gender_df["Price"].mean()

total_purchase = grouped_gender_df["Price"].sum()

average_purchase_total = total_purchase/total_count

#Create the summary table

summary_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_purchase, "Total Purchase Value": total_purchase, "Average Total Purchase per Person": average_purchase_total})

summary_df.index.name = "Gender"

summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:.2f}".format)
summary_df["Total Purchase Value"] = summary_df["Total Purchase Value"].map("${:.2f}".format)
summary_df["Average Total Purchase per Person"] = summary_df["Average Total Purchase per Person"].map("${:.2f}".format)
summary_df.head()
```

```python
#Create bins for ages
age_bins = [0, 9.9, 14.9 , 19.9, 24.9, 29.9, 34.9, 39.9, 999]

# Create the names for the five bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Demographics"] = pd.cut(purchase_data_df["Age"], age_bins, labels=group_names)
purchase_data_df

age_group = purchase_data_df.groupby("Age Demographics")

age_count = age_group["SN"].nunique()

#Find the percentage per aged group
age_percentage = (age_count/total_players) * 100

#Print the summary table
summary_df = pd.DataFrame({"Total Count": age_count, "Percentage of Players":age_percentage})                         
summary_df.index.name = None
summary_df["Percentage of Players"] = summary_df["Percentage of Players"].map("{0:.2f}%".format)                    
summary_df
```

```python
#Create bins for ages
age_bins = [0, 9.9, 14.9 , 19.9, 24.9, 29.9, 34.9, 39.9, 999]

# Create the names for the five bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Demographics"] = pd.cut(purchase_data_df["Age"], age_bins, labels=group_names)
purchase_data_df

age_group = purchase_data_df.groupby("Age Demographics")

#Get the total of purchase per age group
age_purchase_count = age_group["Purchase ID"].count()

#Find the average purchase price per age group
age_average_purchase = age_group["Price"].mean()

#Find the total purchase value per age group
purchase_value = age_group["Price"].sum()

#Find the average total purchase per person
age_average_total_purchase = purchase_value/age_count


#Create the summary table

summary_df = pd.DataFrame({"Purchase Count": age_purchase_count, "Average Purchase Price": age_average_purchase, "Total Purchase Value": purchase_value, "Average Total Purchase per Person": age_average_total_purchase})

summary_df.index.name = "Age Ranges"

summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:.2f}".format)
summary_df["Total Purchase Value"] = summary_df["Total Purchase Value"].map("${:.2f}".format)
summary_df["Average Total Purchase per Person"] = summary_df["Average Total Purchase per Person"].map("${:.2f}".format)
summary_df

```

```python
#Group the top 5 spenders by screen name
top_spenders = purchase_data_df.groupby(['SN'])

#Find the total number of purchases per name
spender_purchase_count = top_spenders["Purchase ID"].count()

#Find the average purchase price per top spender
top_spender_average_purchase = top_spenders["Price"].mean()

#Find the total Purchase Value per top spender
top_spender_total_value = top_spenders["Price"].sum()

#Create the summary table
summary_df = pd.DataFrame({"Purchase Count": spender_purchase_count, "Average Purchase Price": top_spender_average_purchase, "Total Purchase Value": top_spender_total_value})

total_purchase_value_df = summary_df.sort_values("Total Purchase Value", ascending=False)
total_purchase_value_df.head()

total_purchase_value_df["Average Purchase Price"] = total_purchase_value_df["Average Purchase Price"].map("${:.2f}".format)
total_purchase_value_df["Total Purchase Value"] = total_purchase_value_df["Total Purchase Value"].map("${:.2f}".format)
total_purchase_value_df.head()
```

```python
#Retreive the Item ID, Item Name, and Item Price columns
item_titles = purchase_data_df[["Item ID", "Item Name", "Price"]]

#Group data by Item ID & Item Name
item_info = purchase_data_df.groupby(['Item ID', 'Item Name'])

#Find the purchase count of the date
item_count = item_info["Price"].count()

#Find the total purchase value
item_total_value = (item_info["Price"].sum())

#Find the average item price
item_price = item_total_value/item_count

#Create the summary table
summary_df = pd.DataFrame({"Purchase Count": item_count, "Item Price": item_price, "Total Purchase Value": item_total_value})

#Sort the purchase count column into descending order
purchase_count_df = summary_df.sort_values("Purchase Count", ascending=False)
purchase_count_df.head()

purchase_count_df["Item Price"] = purchase_count_df["Item Price"].map("${:.2f}".format)
purchase_count_df["Total Purchase Value"] = purchase_count_df["Total Purchase Value"].map("${:.2f}".format)
purchase_count_df.head()
```
